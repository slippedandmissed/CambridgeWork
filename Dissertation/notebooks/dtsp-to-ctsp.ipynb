```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from tqdm import tqdm

%matplotlib inline
```

```python
mpl.rcParams["font.size"] = 32

fig, ax = plt.subplots(figsize=(20, 18))

ax.set_xlabel("i")
ax.set_ylabel("j")

resolution = 100
i = j = np.linspace(0, 1.5, resolution)
f = np.empty((resolution, resolution))
for jidx, j_ in tqdm(enumerate(j), total=resolution, leave=False):
  for iidx, i_ in enumerate(i):
    f[jidx, iidx] = np.abs(i_ + j_ - 1)

heatmap = ax.pcolor(i, j, f, cmap="gray")
cbar = plt.colorbar(heatmap)

cbar.ax.get_yaxis().set_ticks([])
for j, lab in enumerate(["$\\frac{C(V_i,V_j)}{\\sqrt{2}}$", "$\\uparrow$increasing"]):
    cbar.ax.text(1.5, 2*j, lab, ha="left", va="bottom" if j == 0 else "top")
cbar.ax.get_yaxis().labelpad = 35
cbar.ax.set_ylabel("Cost Density, $f$", rotation=270)

plt.savefig("dtsp-to-ctsp.png")
```
